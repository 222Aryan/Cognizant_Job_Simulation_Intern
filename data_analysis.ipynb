```python
def convert_to_datetime(data: pd.DataFrame = None,  column: str = None):
  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

##SECTION 4 - MERGE DATA

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, "%Y-%m-%d %H:00:00") for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
```

```python
sales_df['product_id'].nunique()
```

```python
stock_df = convert_timestamp_to_hourly(stock_df, "timestamp")
stock_df.head()
```

```python
temp_df = convert_timestamp_to_hourly(temp_df, "timestamp")
temp_df.head()
```

```python
sales_agg = sales_df.groupby(['timestamp','product_id']).agg({'quantity':"sum"}).reset_index()
sales_agg.head(10)
```

```python
stock_agg = stock_df.groupby(['timestamp','product_id']).agg({"estimated_stock_pct":"mean"}).reset_index()
stock_agg.head()
```

```python
temp_agg = temp_df.groupby(["timestamp"]).agg({"temperature": "mean"}).reset_index()
temp_agg.head()
```

```python
merge_df = stock_agg.merge(sales_agg, on=["timestamp","product_id"], how="left")
merge_df.head()
```

```python
merge_df = merge_df.merge(temp_agg, on=["timestamp"], how="left")
merge_df.head()
```

```python
merge_df['quantity'] = merge_df['quantity'].fillna(0)
```

```python
merge_df.info()
```

```python
product_categories = sales_df[["product_id","category"]]
product_categories = product_categories.drop_duplicates()

product_price = sales_df[["product_id","unit_price"]]
product_price = product_price.drop_duplicates()
product_price
```

```python
merge_df = merge_df.merge(product_categories, on ="product_id", how="left")
merge_df.head()
```

```python
merge_df = merge_df.merge(product_price, on="product_id", how="left")
merge_df.head()
```

```python
merge_df.info()
```

##SECTION 5 - FEATURE ENGINERRING


```python
# MODIFYING TIMSTAMP FOR MACHINE LEARNING TRAINING
merge_df["time_day_of_month"] = merge_df["timestamp"].dt.day
merge_df["time_day_of_weak"]   = merge_df["timestamp"].dt.dayofweek
merge_df["time_of_weekday"] = merge_df["timestamp"].dt.weekday
merge_df["timestamp_hour"] = merge_df["timestamp"].dt.hour
merge_df.drop(columns=['timestamp'], inplace=True)
merge_df.head()
```

```python
# MODIFYING CATEGORY BY CONVERTING TO NUMERIC DATA
merge_df = pd.get_dummies(merge_df, columns=['category'], dtype=int)
merge_df.head()
```

```python
merge_df.info()
```

```python
merge_df.drop(columns=['product_id'], inplace=True)
merge_df.head()
```

## SECTION - 7 MODELING

```python
X = merge_df.drop(columns=['estimated_stock_pct'])
y = merge_df['estimated_stock_pct']
print(X.shape)
print(y.shape)
```

```python
K = 10
split = 0.75
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
accuracy = []

for fold in range(0,K):
  model = RandomForestRegressor()
  scaler = StandardScaler()

  X_train,X_test,y_train,y_test = train_test_split(X, y, train_size=split, random_state = 42)

  scaler.fit(X_train)
  X_train = scaler.transform(X_train)
  X_test  = scaler.transform(X_test)

  trained_model = model.fit(X_train,y_train)

  y_pred = trained_model.predict(X_test)

  mae = mean_absolute_error(y_true=y_test,y_pred=y_pred)
  accuracy.append(mae)
  print(f"Fold {fold+1}: MAE = {mae:.3f}")
print(f"AVERAGE MAE: {(sum(accuracy))/len(accuracy):.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
features = [i.split("__")[0] for i in X.columns]
importances = model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title("Feature Importances")
plt.barh(range(len(indices)), importances[indices], color='r', align='center')
plt.yticks(range(len(indices)),[features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
